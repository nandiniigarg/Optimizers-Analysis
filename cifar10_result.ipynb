```python
from keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D
import keras
from tensorflow.keras.optimizers import Adam, Adagrad, Adamax, RMSprop, SGD, Adadelta
import matplotlib.pyplot as plt
import numpy as np
from keras import layers
```

```python
(x_train, y_train), (x_valid, y_valid) = cifar10.load_data()

x_train.shape
```

```python
x_valid.shape
```

```python
print(x_train.min(), x_train.max())
```

```python
x_train = x_train / 255
x_valid = x_valid / 255 
```

```python
x_train = x_train.reshape(50000, 3072)
x_valid = x_valid.reshape(10000, 3072)
num_categories =10
y_train = keras.utils.to_categorical(y_train, num_categories)
y_valid = keras.utils.to_categorical(y_valid, num_categories)
```

```python
model_adam = Sequential()
model_adam.add(Dense(units=128, activation='relu', input_shape=(3072,)))
# model.add(Dense(units = 128, activation='relu'))
model_adam.add(Dense(units = 10, activation='softmax'))
model_adam.summary()

model_adam.compile(loss='categorical_crossentropy', optimizer=Adam(),metrics=['accuracy'])

history_adam = model_adam.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
model_rms_cifar = Sequential()
model_rms_cifar.add(Dense(units=128, activation='relu', input_shape=(3072,)))
# model.add(Dense(units = 128, activation='relu'))
model_rms_cifar.add(Dense(units = 10, activation='softmax'))
model_rms_cifar.summary()

model_rms_cifar.compile(loss='categorical_crossentropy', optimizer=Adam(),metrics=['accuracy'])

histroy_rms_cifar = model_rms_cifar.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
model_adagrad_cifar = Sequential()
model_adagrad_cifar.add(Dense(units=128, activation='relu', input_shape=(3072,)))
# model.add(Dense(units = 128, activation='relu'))
model_adagrad_cifar.add(Dense(units = 10, activation='softmax'))
model_adagrad_cifar.summary()

model_adagrad_cifar.compile(loss='categorical_crossentropy', optimizer=Adagrad(),metrics=['accuracy'])

history_adagrad_cifar = model_adagrad_cifar.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
model_adadelta_cifar = Sequential()
model_adadelta_cifar.add(Dense(units=128, activation='relu', input_shape=(3072,)))
# model.add(Dense(units = 128, activation='relu'))
model_adadelta_cifar.add(Dense(units = 10, activation='softmax'))
model_adadelta_cifar.summary()

model_adadelta_cifar.compile(loss='categorical_crossentropy', optimizer=Adadelta(),metrics=['accuracy'])

history_adadelta_cifar = model_adagrad_cifar.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
model_adamax_cifar = Sequential()
model_adamax_cifar.add(Dense(units=128, activation='relu', input_shape=(3072,)))
# model.add(Dense(units = 128, activation='relu'))
model_adamax_cifar.add(Dense(units = 10, activation='softmax'))
model_adamax_cifar.summary()

model_adamax_cifar.compile(loss='categorical_crossentropy', optimizer=Adamax(),metrics=['accuracy'])

history_adamax_cifar = model_adamax_cifar.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
model_sgd_cifar = Sequential()
model_sgd_cifar.add(Dense(units=128, activation='relu', input_shape=(3072,)))
# model.add(Dense(units = 128, activation='relu'))
model_sgd_cifar.add(Dense(units = 10, activation='softmax'))
model_sgd_cifar.summary()

model_sgd_cifar.compile(loss='categorical_crossentropy', optimizer=SGD(),metrics=['accuracy'])

history_sgd_cifar = model_sgd_cifar.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
model_sgd_cifar = Sequential()
model_sgd_cifar.add(Dense(units=128, activation='relu', input_shape=(3072,)))
# model.add(Dense(units = 128, activation='relu'))
model_sgd_cifar.add(Dense(units = 10, activation='softmax'))
model_sgd_cifar.summary()

model_sgd_cifar.compile(loss='categorical_crossentropy', optimizer=SGD(),metrics=['accuracy'])

history_sgd_cifar = model_sgd_cifar.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
plt.plot(history_adadelta_cifar.history['loss'], label = 'Adadelta TR')
plt.plot(history_adadelta_cifar.history['val_loss'],linestyle="--",label = 'Adadelta TS')

plt.plot(history_sgd_cifar.history['loss'], label = 'SGD TR')
plt.plot(history_sgd_cifar.history['val_loss'], linestyle="--", label = 'SGD TS')

plt.plot(history_adagrad_cifar.history['loss'], label = 'Adagrad TR')
plt.plot(history_adagrad_cifar.history['val_loss'], linestyle="--", label = 'Adagrad TS')
plt.plot(history_adamax_cifar.history['loss'], label = 'Adamax TR')
plt.plot(history_adamax_cifar.history['val_loss'], linestyle="--", label = 'Adamax TS')
plt.plot(histroy_rms_cifar.history['loss'], label = 'RMSprop TR')
plt.plot(histroy_rms_cifar.history['val_loss'], linestyle="--",  label = 'RMSprop TS')
plt.plot(history_adam.history['loss'], label = 'Adam TR')
plt.plot(history_adam.history['val_loss'], linestyle="--", label ='Adam TS')
plt.title('CIFAR10 Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.figure(figsize=(18, 12))
# plt.savefig('cifar10_result.png')
plt.show()
```

```python
plt.plot(history_adadelta_cifar.history['accuracy'], label = 'Adadelta TR')
plt.plot(history_adadelta_cifar.history['val_accuracy'],linestyle="--",label = 'Adadelta VL')

plt.plot(history_sgd_cifar.history['accuracy'], label = 'SGD TR')
plt.plot(history_sgd_cifar.history['val_accuracy'], linestyle="--", label = 'SGD VL')

plt.plot(history_adagrad_cifar.history['accuracy'], label = 'Adagrad TR')
plt.plot(history_adagrad_cifar.history['val_accuracy'], linestyle="--", label = 'Adagrad VL')

plt.plot(history_adamax_cifar.history['accuracy'], label = 'Adamax TR')
plt.plot(history_adamax_cifar.history['val_accuracy'], linestyle="--", label = 'Adamax VL')

plt.plot(histroy_rms_cifar.history['accuracy'], label = 'RMSprop TR')
plt.plot(histroy_rms_cifar.history['val_accuracy'], linestyle="--",  label = 'RMSprop VL')

plt.plot(history_adam.history['accuracy'], label = 'Adam TR')
plt.plot(history_adam.history['val_accuracy'], linestyle="--", label ='Adam VL')

plt.title('CIFAR10 Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.figure(figsize=(18, 12))
# plt.savefig('caltech_result.png')
plt.show()
```
