## Fashion MNIST Dataset

```python
from keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D
import keras
from tensorflow.keras.optimizers import Adam, Adagrad, Adamax, RMSprop, SGD, Adadelta
import matplotlib.pyplot as plt
import numpy as np
from keras import layers
```

```python
(x_train, y_train), (x_valid, y_valid) = fashion_mnist.load_data()

x_train.shape
```

```python
# Y = data_train['label'].values
# X = data_train.drop(columns=['label']).values/255.0
```

```python
x_valid.shape
```

```python
print(x_train.min(), x_train.max())
```

Normalizing

```python
x_train = x_train / 255
x_valid = x_valid / 255 
```

```python
print(x_train.min(), x_train.max())
```

```python
x_train = x_train.reshape(60000, 784)
x_valid = x_valid.reshape(10000, 784)

num_categories = 10

y_train = keras.utils.to_categorical(y_train, num_categories)
y_valid = keras.utils.to_categorical(y_valid, num_categories)
```

```python
# Adam Optiimizer for Fashion MNIST

from tensorflow.keras.layers import Dense
model_adam_fmnist = Sequential()
model_adam_fmnist.add(Dense(units=128, activation='relu', input_shape=(784,)))
# model.add(Dense(units = 128, activation='relu'))
model_adam_fmnist.add(Dense(units = 10, activation='softmax'))
model_adam_fmnist.summary()

model_adam_fmnist.compile(loss='categorical_crossentropy', optimizer=Adam(),metrics=['accuracy'])

history_adam_fmnist = model_adam_fmnist.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
# RMSprop Optimizer for Fashion MNIST

model_rms_fmnist = Sequential()
model_rms_fmnist.add(Dense(units=128, activation='relu', input_shape=(784,)))
model_rms_fmnist.add(Dense(units = 10, activation='softmax'))
model_rms_fmnist.summary()

model_rms_fmnist.compile(loss='categorical_crossentropy', optimizer=RMSprop(),metrics=['accuracy'])

history_rms_fmnist = model_rms_fmnist.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
# Adagrad Optiimizer for Fashion MNIST

model_adagrad_fmnist = Sequential()
model_adagrad_fmnist.add(Dense(units=128, activation='relu', input_shape=(784,)))
model_adagrad_fmnist.add(Dense(units = 10, activation='softmax'))
model_adagrad_fmnist.summary()

model_adagrad_fmnist.compile(loss='categorical_crossentropy', optimizer=Adagrad(),metrics=['accuracy'])

history_adagrad_fmnist = model_adagrad_fmnist.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
# Adamax Optiimizer for Fashion MNIST

model_adamax_fmnist = Sequential()
model_adamax_fmnist.add(Dense(units=128, activation='relu', input_shape=(784,)))
model_adamax_fmnist.add(Dense(units = 10, activation='softmax'))
model_adamax_fmnist.summary()

model_adamax_fmnist.compile(loss='categorical_crossentropy', optimizer=Adamax(),metrics=['accuracy'])

history_adamax_fmnist = model_adamax_fmnist.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
# SGD Optiimizer for Fashion MNIST

model_sgd_fmnist = Sequential()
model_sgd_fmnist.add(Dense(units=128, activation='relu', input_shape=(784,)))
model_sgd_fmnist.add(Dense(units = 10, activation='softmax'))
model_sgd_fmnist.summary()

model_sgd_fmnist.compile(loss='categorical_crossentropy', optimizer=SGD(),metrics=['accuracy'])

history_sgd_fmnist = model_sgd_fmnist.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
# Adadelta Optiimizer for Fashion MNIST

model_adadelta_fmnist = Sequential()
model_adadelta_fmnist.add(Dense(units=128, activation='relu', input_shape=(784,)))
model_adadelta_fmnist.add(Dense(units = 10, activation='softmax'))
model_adadelta_fmnist.summary()

model_adadelta_fmnist.compile(loss='categorical_crossentropy', optimizer=Adadelta(),metrics=['accuracy'])

history_adadelta_fmnist = model_adadelta_fmnist.fit(
    x_train, y_train, epochs=30, verbose=1, validation_data=(x_valid, y_valid)
)
```

```python
plt.plot(history_adadelta_fmnist.history['loss'], label = 'Adadelta TR')
plt.plot(history_adadelta_fmnist.history['val_loss'],linestyle="--",label = 'Adadelta VL')

plt.plot(history_sgd_fmnist.history['loss'], label = 'SGD TR')
plt.plot(history_sgd_fmnist.history['val_loss'], linestyle="--", label = 'SGD VL')

plt.plot(history_adagrad_fmnist.history['loss'], label = 'Adagrad TR')
plt.plot(history_adagrad_fmnist.history['val_loss'], linestyle="--", label = 'Adagrad VL')

plt.plot(history_adamax_fmnist.history['loss'], label = 'Adamax TR')
plt.plot(history_adamax_fmnist.history['val_loss'], linestyle="--", label = 'Adamax VL')

plt.plot(history_rms_fmnist.history['loss'], label = 'RMSprop TR')
plt.plot(history_rms_fmnist.history['val_loss'], linestyle="--",  label = 'RMSprop VL')

plt.plot(history_adam_fmnist.history['loss'], label = 'Adam TR')
plt.plot(history_adam_fmnist.history['val_loss'], linestyle="--", label ='Adam VL')

plt.title('Fashion MNIST Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.figure(figsize=(18, 12))
plt.savefig('fmnist_result')
plt.show()
```

```python
plt.plot(history_adadelta_fmnist.history['accuracy'], label = 'Adadelta TR')
plt.plot(history_adadelta_fmnist.history['val_accuracy'],linestyle="--",label = 'Adadelta VL')

plt.plot(history_sgd_fmnist.history['accuracy'], label = 'SGD TR')
plt.plot(history_sgd_fmnist.history['val_accuracy'], linestyle="--", label = 'SGD VL')

plt.plot(history_adagrad_fmnist.history['accuracy'], label = 'Adagrad TR')
plt.plot(history_adagrad_fmnist.history['val_accuracy'], linestyle="--", label = 'Adagrad VL')

plt.plot(history_adamax_fmnist.history['accuracy'], label = 'Adamax TR')
plt.plot(history_adamax_fmnist.history['val_accuracy'], linestyle="--", label = 'Adamax VL')

plt.plot(history_rms_fmnist.history['accuracy'], label = 'RMSprop TR')
plt.plot(history_rms_fmnist.history['val_accuracy'], linestyle="--",  label = 'RMSprop VL')

plt.plot(history_adam_fmnist.history['accuracy'], label = 'Adam TR')
plt.plot(history_adam_fmnist.history['val_accuracy'], linestyle="--", label ='Adam VL')

plt.title('Fashion MNIST Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.figure(figsize=(18, 12))
# plt.savefig('caltech_result.png')
plt.show()
```
